```python
i=c.shape[0]
TF=np.zeros(shape=(i,i))
for i in range(0,c.shape[0]):
  for j in range(0,c.shape[1]):
    TF[i,j]=c[i,i]+c[j,j]
TF=TF - np.diag(np.diag(TF))#matrix off_diagonal


print(TF)
TFdf = pd.DataFrame(TF)
store = pd.HDFStore('/content/drive/MyDrive/thesis_data/TenTFdf.h5')
store['TFdf'] =TFdf
store.close()
```

```python
#while(False):
i=up_adj_mtx.shape[0]
jaccard_similarity=np.zeros(shape=(i,i))
for i in range(0,jaccard_similarity.shape[0]):
  for j in range(0,jaccard_similarity.shape[1]):
    if CF[i,j]==0 or TF[i,j]==0:
      jaccard_similarity[i,j]==0
    else:

      #jaccard_similarity[i,j]=(CF[i,j]/TF[i,j])
      jaccard_similarity[i,j]=jnp.divide(CF[i,j],TF[i,j]).block_until_ready()

jaccard_similarityR=jaccard_similarity - np.diag(np.diag(jaccard_similarity))#matrix off_diagonal without rounding      
#jaccard_similarityR=jaccard_similarityR - np.diag(np.diag(jaccard_similarityR))#matrix off_diagonal
jaccard_similaritydfR = pd.DataFrame(jaccard_similarityR)
store = pd.HDFStore('/content/drive/MyDrive/thesis_data/Tenjaccard_similaritydfR.h5')
store['jaccard_similaritydfR'] = jaccard_similaritydfR
print(np.round(jaccard_similarityR,1))
```
